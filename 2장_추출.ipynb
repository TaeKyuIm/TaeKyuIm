# 조건에 따른 데이터 행 추출

조건에 따른 추출은 특정 데이터에 조건을 걸어 특정 데이터 만을 얻어내는 것입니다.

```python
reserve_tb[(reserve_tb['checkout_date'] >= '2016-10-13') & (reserve_tb['checkout_date'] <= '2016-10-14')]
# 체크아웃 날짜를 기준으로 2016-10-13 에서 2016-10-14 사이에 체크아웃한 날짜에 해당되는 인덱스만을 추출했습니다.
```

```python
reserve_tb = pd.read_csv('reserve.csv', encoding='UTF-8')
reserve_tb.loc[(reserve_tb['checkout_date'] >= '2016-10-13') & (reserve_tb['checkout_date'] <= '2016-10-14'), :]
# 위에 한번 나왔던것 처럼 loc 에 콤마쌍을 넣으면 인덱스도 뽑아낼 수 있습니다.
```

```python
reserve_tb.query('"2016-10-13" <= checkout_date <= "2016-10-14"')
# pandas의 설명상 query()는 Query the columns of a DataFrame with a boolean expression.
# 즉 참-거짓 표현으로 column들을 추출하는 함수라고 소개되어있습니다.
# pandas help 사이트에서 보시면 ()안에 문자열 형태로 참 거짓 문을 넣으시면 됩니다.
```

# 데이터 값을 고려하지 않는 샘플링

데이터 값이 너무 많은 경우 샘플링으로 데이터 값을 줄일 수 있습니다.
샘플링의 경우 사용자가 조건을 선언할 수 있지만, 보통의 경우 랜덤으로 샘플링을 합니다.

```python
reserve_tb.sample(frac=0.5)
# sample()은 행단위의 샘플링 함수 입니다. 위의 구문처럼 frac을 쓸 경우 전체에서 비율에 맞게 샘플이 되고,
# sample(n = 1000)처럼 작성할경우 특정 개수만큼 샘플리을 진행하게 됩니다.
```

# 집약 ID에 기반한 샘플링

샘플링의 경우 공평한 샘플링 구현이 중요합니다. 공평한 샘플링이란 말이 이해가 안될 수 있는데,
분석 대상의 단위와 샘플링의 단위를 맞추어야 합니다.
즉 설명을 드리자면, 해당 자료는 customer_id가 여러개로 중복되어 있습니다.
즉 분석 대상은 사람 한명인데, 샘플링은 두번이므로 어떤 사람은 두번 제거되고, 어떤 사람은 한번도 제거가 안될 수 있습니다.
즉 중복되는 자료들을 정리하여 샘플링을 하여야 공평하게 샘플링을 할 수 있습니다.

```python
target = pd.Series(reserve_tb['customer_id'].unique()).sample(frac = 0.5)
# unique()함수는 Series에 대해 중복을 제거한 Series를 반환합니다.
# 즉 custome_id에 대해 중복을 제거하고(unique()함수) 샘플링(sample())을 진행할 경우 아이디에 대해서만
# 공평하게 샘플링을 할 수 있습니다.
```

```python
reserve_tb[reserve_tb['customer_id'].isin(target)]
# 해당 함수 isin()은 Series, DataFrame, dictionary를 넣을경우 해당되는 값이 있을경우 true, 없을경우 false 를 반환합니다.
# 즉 true가 반환할경우 우리가 샘플링한 id가 선택되므로, 선택되는 id가 있는 column을 찾아 행을 추출합니다.
```
